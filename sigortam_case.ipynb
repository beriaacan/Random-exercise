1. Ortalama bir kasko poliçe tutarı ne kadardır?


Her bir segment için "Gross Revenue" (Brüt Gelir) ile "Sales" (Gerçek Müşteri) sayısının bölünmesiyle bulunabilir.


$$
\text{Average Ticket Size} = \frac{\text{Gross Revenue}}{\text{Sales}}
$$


```python
def calculate_average_policy_amount(sales, gross_revenue):
    return gross_revenue / sales
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    sales = row['Sales']
    gross_revenue = row['Gross Revenue  (₺)']
    total_sales = df.loc['Total', 'Sales']


    average_policy_amount = calculate_average_policy_amount(sales, gross_revenue)


    print(f"{segment} Segmenti için Ortalama Kasko Poliçe Tutarı: {average_policy_amount:.2f} ₺")


total_gross_revenue = df.loc['Total', 'Gross Revenue  (₺)']
average_policy_amount_total = calculate_average_policy_amount(total_sales, total_gross_revenue)
print(f"Ortalama Kasko Poliçe Tutarı: {average_policy_amount_total:.2f} ₺")
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    sales = row['Sales']
    gross_revenue = row['Gross Revenue  (₺)']
    total_sales = df.loc['Total', 'Sales']

    average_policy_amount = calculate_average_policy_amount(sales, gross_revenue)


    df.at[index, 'Ticket Size'] = average_policy_amount


df.loc['Total'] = df.iloc[:, 1:].sum()
df
```

3. Yapılan satışlardaki indirim oranı nedir?



$$
\text{Discount Rate} = \frac{\text{Discount}}{\text{Gross Revenue}}

```python
def calculate_discount_rate(discount, gross_revenue):
    return discount / gross_revenue
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    discount = row['Discount (₺)']
    gross_revenue = row['Gross Revenue  (₺)']

    discount_rate = calculate_discount_rate(discount, gross_revenue)

    print(f"{segment} Segmenti için İndirim Oranı: {discount_rate:.4f}")
```

4. 1 yılın sonundaki net kazanç nedir?

$$
\text{Net Kâr (1 Yıl)} = \text{Satış} \times \text{Brüt Gelir} \times (1 - \text{İndirim Oranı}) \times 0.2
$$

```python
def calculate_profit_1_year(sales, gross_revenue, discount_rate):
    commission_rate = 0.2
    return sales * gross_revenue * (1 - discount_rate) * commission_rate
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    sales = row['Sales']
    gross_revenue = row['Gross Revenue  (₺)']
    discount_rate = row['Discount Rate']

    profit_1_year = calculate_profit_1_year(sales, gross_revenue, discount_rate)

    print(f"{segment} Segmenti için 1 Yılın Sonundaki Net Kazanç: {profit_1_year:.2f} ₺")


```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    sales = row['Sales']
    gross_revenue = row['Gross Revenue  (₺)']
    discount_rate = row['Discount Rate']

    profit_1_year = calculate_profit_1_year(sales, gross_revenue, discount_rate)

    df.at[index, 'Profit (1 year)'] = profit_1_year

df.loc['Total'] = df.iloc[:, 1:].sum()
df
```

5. 3 yılın sonundaki toplam net kazanç nedir?


$$
\text{Net Kâr (1 Yıl)} = \text{Satış} \times \text{Brüt Gelir} \times (1 - \text{İndirim Oranı}) \times 0.2
$$

$$
\text{Net Kâr (3 Yıl)} = \text{Net Kâr (1 Yıl)} \times 3
$$

```python
def calculate_total_profit_3_years(profit_1_year):
    return profit_1_year * 3
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    profit_1_year = row['Profit (1 year)']

    total_profit_3_years = calculate_total_profit_3_years(profit_1_year)

    print(f"{segment} Segmenti için 3 Yılın Sonundaki Toplam Net Kazanç: {total_profit_3_years:.2f} ₺")



```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    profit_1_year = row['Profit (1 year)']

    total_profit_3_years = calculate_total_profit_3_years(profit_1_year)

    df.at[index, 'Total Profit (3 years)'] = total_profit_3_years

df.loc['Total'] = df.iloc[:, 1:].sum()
df
```
