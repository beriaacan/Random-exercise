1. Ortalama bir kasko poliçe tutarı ne kadardır?

$$
\text{Average Ticket Size} = \frac{\text{Gross Revenue}}{\text{Sales}}
$$


```python
def calculate_average_policy_amount(sales, gross_revenue):
    return gross_revenue / sales
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    sales = row['Sales']
    gross_revenue = row['Gross Revenue  (₺)']
    total_sales = df.loc['Total', 'Sales']


    average_policy_amount = calculate_average_policy_amount(sales, gross_revenue)


    print(f"{segment} Segmenti için Ortalama Kasko Poliçe Tutarı: {average_policy_amount:.2f} ₺")


total_gross_revenue = df.loc['Total', 'Gross Revenue  (₺)']
average_policy_amount_total = calculate_average_policy_amount(total_sales, total_gross_revenue)
print(f"Ortalama Kasko Poliçe Tutarı: {average_policy_amount_total:.2f} ₺")
```

2. Satışa dönüşme oranı nedir?

Satışa dönüşme oranı, potansiyel müşterilerin (lead) gerçek müşterilere (sales) dönüşme oranını ifade eder.


$$
\text{Conversion Rate} = \frac{\text{Sales}}{\text{Lead}}
$$

```python
def calculate_conversion_rate(sales, lead):
    return sales / lead
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    sales = row['Sales']
    lead = row['Lead']

    conversion_rate = calculate_conversion_rate(sales, lead)

    print(f"{segment} Segmenti için Satışa Dönüşme Oranı: {conversion_rate:.4f}")

# Calculate and print total conversion rate
total_sales = df['Sales'].sum()
total_lead = df['Lead'].sum()
total_conversion_rate = calculate_conversion_rate(total_sales, total_lead)

print(f"\nToplam Satışa Dönüşme Oranı: {total_conversion_rate:.4f}\n")
```

```python
def calculate_conversion_rate(initial_rate, annual_increase_rate, years):
    conversion_rates = [initial_rate]

    for year in range(1, years):
        initial_rate *= (1 + annual_increase_rate)
        conversion_rates.append(initial_rate)

    return conversion_rates

# Başlangıç dönüşüm oranları
initial_rates = {
    'A': 0.0250,
    'B': 0.0175,
    'C': 0.0080,
    'D': 0.0250
}

annual_increase_rate = 0.15
years = 4

conversion_rates_result = {}
for segment, initial_rate in initial_rates.items():
    conversion_rates_result[segment] = calculate_conversion_rate(initial_rate, annual_increase_rate, years)

total_conversion_rates = [sum(rates) / len(rates) for rates in zip(*conversion_rates_result.values())]

for segment, rates in conversion_rates_result.items():
    print(f"{segment} Segmenti Dönüşüm Oranları (Conversion Rate):")
    for year, rate in enumerate(rates, start=1):
        print(f"Yıl {2022 + year}: {rate:.6f}")
    print()

print("Toplam Dönüşüm Oranları (Total Conversion Rate):")
for year, total_rate in enumerate(total_conversion_rates, start=1):
    print(f"Yıl {2022 + year}: {total_rate:.6f}")
```

4. 1 yılın sonundaki net kazanç nedir?

$$
\text{Net Kazanç (1 Yıl)} = (\text{Gross Revenue} \times \text{ Komisyon oranı} )-(\text{Discount} + \text{ Marketing} )
$$

```python

def calculate_profit_1_year(gross_revenue, discount, marketing, commission_rate=0.2):
    return (gross_revenue * commission_rate) - (discount + marketing)
```

```python
for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    gross_revenue = row['Gross Revenue  (₺)']
    discount = row['Discount (₺)']
    marketing = row['Marketing']


    profit_1_year = calculate_profit_1_year(gross_revenue, discount, marketing)


    print(f"{segment} Segmenti için 1 Yıl Sonundaki Net Kazanç: {profit_1_year:.2f} ₺")


    df.at[index, 'Profit (1 year)'] = profit_1_year


total_profit_1_year = df.loc['Total', 'Profit (1 year)']
print(f"\nToplam 1 Yıl Sonundaki Net Kazanç: {total_profit_1_year:.2f} ₺\n")

df.loc['Total', 'Profit (1 year)'] = df.iloc[:-1]['Profit (1 year)'].sum()
df
```

5. 3 yılın sonundaki toplam net kazanç nedir?

3 Yılın Sonundaki Toplam Net Kazanç:

Bu, yıllık net kazançların toplamıdır. İkinci ve üçüncü yıllarda, ticket size'ın %20 artacağını ve marketing harcamalarının olmayacağını dikkate almalıyız. Conversion rate'in de her yıl %15 artacağını unutmamalıyız.


$$
\text{Net Kâr (1 Yıl)} = \text{Satış} \times \text{Brüt Gelir} \times (1 - \text{İndirim Oranı}) \times 0.2
$$

$$
\text{Net Kâr (3 Yıl)} = \text{Net Kâr (1 Yıl)} \times 3
$$

```python
def calculate_net_profit(gross_revenue, discount, marketing, commission_rate=0.2):
    return (gross_revenue * commission_rate) - (discount + marketing)
```

```python
def calculate_total_net_profit_3_years(net_profit_1_year, ticket_size_increase_rate, conversion_rate_increase_rate, years):
    total_net_profit_3_years = net_profit_1_year * years

    for year in range(2, years + 1):
        net_profit_1_year *= (1 + ticket_size_increase_rate) * (1 + conversion_rate_increase_rate)
        total_net_profit_3_years += net_profit_1_year

    return total_net_profit_3_years
```

```python
ticket_size_increase_rate = 0.20
conversion_rate_increase_rate = 0.15

for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    gross_revenue = row['Gross Revenue  (₺)']
    discount = row['Discount (₺)']
    marketing = row['Marketing']


    net_profit_1_year = calculate_net_profit(gross_revenue, discount, marketing)

    total_net_profit_3_years = calculate_total_net_profit_3_years(net_profit_1_year, ticket_size_increase_rate, conversion_rate_increase_rate, years=3)

    print(f"{segment} Segmenti için 3 Yıl Sonundaki Toplam Net Kazanç: {total_net_profit_3_years:.2f} ₺")

```

```python

total_net_profit_3_years = df.loc['Total', 'Total Profit (3 years)']
print(f"\nToplam 3 Yıl Sonundaki Toplam Net Kazanç: {total_net_profit_3_years:.2f} ₺\n")


for index, row in df.iterrows():
    if index == 'Total':
        continue

    segment = row['Segment']
    gross_revenue = row['Gross Revenue  (₺)']
    discount = row['Discount (₺)']
    marketing = row['Marketing']

    net_profit_1_year = calculate_net_profit(gross_revenue, discount, marketing)

    total_net_profit_3_years = calculate_total_net_profit_3_years(net_profit_1_year, ticket_size_increase_rate, conversion_rate_increase_rate, years=3)

    df.at[index, 'Total Profit (3 years)'] = total_net_profit_3_years

df.loc['Total', 'Total Profit (3 years)'] = df.iloc[:-1]['Total Profit (3 years)'].sum()
df
```
